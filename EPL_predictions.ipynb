```python
import pandas as pd 
import requests 

API_TOKEN="b877382003f2452488370a1617494dc8"
headers = {"X-Auth-Token":API_TOKEN}
url= "https://api.football-data.org/v4/competitions/PL/matches?season=2024"
response = requests.get(url, headers=headers)
data= response.json()
data.keys()
```

```python
matches=pd.DataFrame(data["matches"])
matches["homeTeam"]=matches["homeTeam"].apply(lambda x:x["name"])
matches["awayTeam"]=matches["awayTeam"].apply(lambda x:x["name"])
matches["score"]=matches["score"].apply(lambda x:x["winner"])
matches
```

```python
home_wins=matches[matches["score"]=="HOME_TEAM"].groupby("homeTeam").size()
away_wins=matches[matches["score"]=="AWAY_TEAM"].groupby("awayTeam").size()
total_wins=(home_wins+away_wins).sort_values(ascending=False)
total_wins
```
